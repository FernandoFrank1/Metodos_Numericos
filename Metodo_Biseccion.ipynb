```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    #return np.sin(2*x+1)-3*x/5
    return np.exp(-x)-x
def bisec(f,a,b,tol=0.001,n=100):
    for i in range(n):
        x=(a+b)/2
        if f(x)*f(b)<0:
            a=x
        elif f(x)*f(a)<0:
            b=x
        else:
            break
        xa=x
        x=(a+b)/2
        ea=abs(x-xa)
        er=abs((x-xa)/x)
        if ea<tol:
            break
    print(i,x,a,b,f(x),'{:%}'.format (ea),'{:%}'.format (er))
bisec(f,0,2)
#Gragica
x=np.linspace(0,4,100)
plt.plot(x,f(x))
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Método de la bisección')
plt.show
```

```python
def polinomio(x):
    # Define el polinomio: x^3 - x^2 + 2x + 1
    return x**3 - x**2 + 2*x + 1

def derivada_polinomio(x):
    # Calcula la derivada del polinomio: 3x^2 - 2x + 2
    return 3*x**2 - 2*x + 2

def newton_raphson(polinomio, derivada_polinomio, x_inicial, tolerancia=1e-6, max_iter=100):
    x = x_inicial
    for _ in range(max_iter):
        fx = polinomio(x)
        if abs(fx) < tolerancia:
            return x
        dfx = derivada_polinomio(x)
        if dfx == 0:
            break
        x = x - fx / dfx
    return None

# Calcula las raíces reales del polinomio
raices = []
for x_inicial in range(-10, 10):
    raiz = newton_raphson(polinomio, derivada_polinomio, x_inicial)
    if raiz is not None:
        raices.append(raiz)

print("Las raíces reales del polinomio son:\n", raices)

```
